```python
# Run the following code to import the dataset to the df (dataframe)
# Do not change this cell!
%matplotlib inline
import pandas as pd
import requests
import io

# Read the data file (csv format)
f = open("flag.data", "r")
csv = f.read() 

# Create a Pandas dataframe from csv data
df = pd.read_csv(io.StringIO(csv), header=None)
```

```python
# Question 2: This dataset does not contain any labels. The list bellow 
# contains the labels in the correct order. Find a way to add them to the dataframe,
# and then print the first few rows again to show the labels.
labels = ['name', 'landmass', 'zone', 'area', 'population', 'language', 'religion', 'bars', 'stripes', 'colours', 'red', 
          'green', 'blue', 'gold', 'white', 'black', 'orange', 'mainhue', 'circles', 'crosses', 'saltires', 'quarters', 
          'sunstars', 'crescent', 'triangle', 'icon', 'animate', 'text', 'topleft', 'botright']

number_of_columns = len(df.columns)

# Change all the labels for the columns. 
for iterator in range(number_of_columns):
    df.rename(columns={iterator: labels[iterator]}, inplace=True)

df
```

```python
# Question 3: Check if the dataset has any NaNs, and if so, show the lines that do.
df_has_any_nan = df.isnull().values.any()

if df_has_any_nan:
    print("Dataframe has some NAN values")
    df[df.isna().any(axis=1)]
else:
    print("Dataframe does not have any NAN values")
    
```

```python
# Question 4: As you can see in the link provided above, the landmass label has 6 values, that correspond to:
# 1=N.America, 2=S.America, 3=Europe, 4=Africa, 5=Asia, 6=Oceania. Replace the landmass column values, 
# with their names from the list above

landmass_labels = ['N.America', 'S.America', 'Europe', 'Africa', 'Asia', 'Oceania']

for iterator in range(6):
    df["landmass"].replace({(iterator+1): landmass_labels[iterator]}, inplace=True)
    
df
```

```python
# Question 5: The area is in 'in thousands of square km'. Transform it in square km.

df["area"] = 1000 * df["area"]

df
```

```python
# Questions 6-9: Find all the flags that:
# - have a population of more than 100 million people (inclusive)

print("Countries that have a population of more than 100 million people (inclusive):")
print(df[df.population >= 100].name)

# - have an area of less than 1000 square km
print("Countries that have an area of less than 1000 square km:")
print(df[df.area < 1000].name)

# - have NO bars and NO stripes
print("Countries that NO bars and NO stripes in their flags:")
print(df[(df.bars == 0) & (df.stripes == 0)].name)

# - are in Oceania
print("Countries that are in Oceania:")
print(df[df.landmass == "Oceania"].name)
```

#### Project

Imagine we need to use this dataset for some machine learning purposes. And in the real world, 80% of the job is cleaning and transforming the data.

We need to create a way to show which color combinations are the most common. To simplify it, we will do it just in pairs, so we will only check the colors by 2 e.g red-white, red-blue, red-gold, etc. Then we will find the top 5 combinations and plot them in a bar chart.

##### General instructions:
 - Create a new dataframe keeping only the color labels.
 - Add new columns in this dataframe by combining the features by 2. DO NOT duplicate the data, e.g red-gold is the same as gold-red, so you need to check for this when creating the dataframe
 - If a combination exists in the dataframe, add 1 to the combination column.
 - Plot the top 5 combinations.
 
*Important! Do not do the pairs by hand, as in most cases you can have datasets of hundreds of features so this is not feasible. Automate the procedure of pairing, and checking if the pair exists.*


```python
colours = ['red', 'green', 'blue', 'gold', 'white', 'black', 'orange']

# Create a new dataframe with a sub-section of the previous one. 
df_colours = df[colours].copy()
df_colours

# Create new columns, based on combinations. 
# Prepare the required text format.
text_template = "{}-{}"

# Initialize colour combinations list.
colour_combination_list = []

# Traverse all elements for the first colour. 
for first_colour_index in range (len(colours)):
    first_colour = colours[first_colour_index]
    
    # Take only the next colours (starting from 'firstColourIndex+1') to create unique colour combinations. 
    for second_colour_index in range(first_colour_index+1, len(colours)):
        second_colour = colours[second_colour_index]
        
        # Create a colour combination named 'colour_combination'
        colour_combination = text_template.format(first_colour, second_colour)
        
        # Update colour combination list. 
        colour_combination_list.append(colour_combination)
        
        # Perform a logic 'AND' operation and write it on a new column named 'colour_combination'
        df_colours[colour_combination] = df_colours[first_colour] * df_colours[second_colour]
        
# Get the columns with labels that are in the "colour_combination_list",
# filter only the "1" rows, 
# count the number of '1' values and sort the combinations based on these count values, 
# take the top 5 count value rows 
# plot a bar chart
plot_object = df_colours[df_colours[colour_combination_list]==1].count().sort_values(ascending=False).head(5).plot(kind='bar', title='Most frequent colour combinations counts of the countries')

plot_object.set_xlabel('Number of countries with the same colour combinations')
plot_object.set_ylabel('Number of same colour combinations among countries')


```
